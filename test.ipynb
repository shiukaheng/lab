```python
from rrt_star import *
import matplotlib.pyplot as plt
```

```python
r = RTTStar(
    2,
    initial=np.array([-50, 0]),
    lower_bounds=np.array([-100, -100]),
    upper_bounds=np.array([100, 100]),
    step_size=2,
    neighbor_radius=10,
    collision_samples=100,
    goal=np.array([50, 50]),
    goal_radius=0.1,
)
```

```python
r.solve(verbose=True, post_goal_iterations=5000)
```

```python
for node in r.nodes:
    if node.parent is not None:
        plt.plot([node.parent.point[0], node.point[0]], [node.parent.point[1], node.point[1]], 'k-')
        # plt.scatter(node.point[0], node.point[1], c='b')
plt.show()
```
