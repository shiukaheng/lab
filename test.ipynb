```python
from rtt_star import *
import matplotlib.pyplot as plt
```

```python
r = RTTStar(
    2,
    np.array([0, 0]),
    lower_bounds=np.array([-100, -100]),
    upper_bounds=np.array([100, 100]),
    step_size=2,
    neighbor_radius=10,
    collision_samples=100
)
```

```python
i = 0
for _ in range(10000):
    i += 1
    r.sample()
```

```python
for node in r.nodes:
    if node.parent is not None:
        plt.plot([node.parent.point[0], node.point[0]], [node.parent.point[1], node.point[1]], 'k-')
plt.show()
```
